## 2) Tracé des vecteurs vitesses 

<div class="alert alert-warning">
<b>Travail à faire : Exécuter le programme suivant afin de tracer les vecteurs vitesses de la balle au cours de son mouvement 

```python
import matplotlib.pyplot as plt
from scipy import * # pour la fonction sqrt

# les listes t, x et y 

t = [ 0.04, 0.08, 0.12, 0.16, 0.2, 0.24, 0.28, 0.32, 0.36, 0.4, 0.44, 0.48]

x = [0.294463, 0.25374, 0.231811, 0.222414, 0.225546, 0.247474, 0.263137, 0.263137, 0.26627, 0.281933, 0.310126, 0.319524]

y = [0.0783147, 0.206751, 0.36338, 0.560733, 0.742423, 0.902185, 1.19978, 1.36581, 1.48485, 1.60702, 1.71979, 1.90148]

# on trace la trajectoire 
plt.axis('equal')
plt.title("trajectoire d'un balle lancée verticalement vers le haut ")
plt.ylabel("y en m")
plt.xlabel("x en m ")
plt.grid()
plt.plot(x,y,'or')

## Représentation du vecteur vitesse 
vx=[0]
vy=[0]

for i in range(1,len(x)-1):
    vx.append((x[i+1]-x[i-1])/(t[i+1]-t[i-1])) # Le calcul est effectué ici entre deux points entourant la position considérée
    vy.append((y[i+1]-y[i-1])/(t[i+1]-t[i-1]))
    print("v=",sqrt(vx[i]**2+vy[i]**2), "m/s")

plt.quiver(x[1:-1],y[1:-1],vx,vy,color='blue',width=0.003,scale=1/0.01) # 'scale' est l'échelle
plt.legend(['Position','Vitesse'])
plt.show()
```

# 4) Pour aller plus loin : Comment évolue  la vitesse au cours du mouvement? 

<div class="alert alert-warning">
<b>Travail à faire : En utilisant le programme du 2) écrire ci dessous les lignes de codes permettant d'afficher le graphe v en fonction du temps. 

```python
## programme permettant d'afficher le graphe vitesse en fonction du temps.
import matplotlib.pyplot as plt
from scipy import * # pour la fonction sqrt

# les listes t, x et y 

t = [ 0.04, 0.08, 0.12, 0.16, 0.2, 0.24, 0.28, 0.32, 0.36, 0.4, 0.44, 0.48]

x = [0.294463, 0.25374, 0.231811, 0.222414, 0.225546, 0.247474, 0.263137, 0.263137, 0.26627, 0.281933, 0.310126, 0.319524]

y = [0.0783147, 0.206751, 0.36338, 0.560733, 0.742423, 0.902185, 1.19978, 1.36581, 1.48485, 1.60702, 1.71979, 1.90148]

## Représentation du vecteur vitesse 
vx=[0]
vy=[0]
v=[]


for i in range(1,len(x)-1):
    vx.append((x[i+1]-x[i-1])/(t[i+1]-t[i-1])) # Le calcul est effectué ici entre deux points entourant la position considérée
    vy.append((y[i+1]-y[i-1])/(t[i+1]-t[i-1]))
    v.append(sqrt((vx[i]**2+vy[i]**2)))
  
plt.xlabel("Temps (en s)")
plt.ylabel("v en m/s")
plt.title("évolution de la coordonnée selon l'axe y du vecteur vitesse ")
plt.grid()   # présence d'une grille
plt.plot(0,0)
plt.plot(t[1:-1],v, "k+") 
plt.show()
```
